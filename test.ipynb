```python
import base64

image_url = "data/21020_hieu.jpg"
encoded_string = base64.b64encode(open(image_url, "rb").read()).decode('utf-8')
print(encoded_string)
```

```python
import anthropic

client = anthropic.Anthropic()
message = client.messages.create(
    model="claude-3-7-sonnet-20250219",
    max_tokens=2048,
    temperature=0.1,
    system=[
      {
        "type": "text",
        "text": open("mvc/controller/prompt.txt", "r", encoding="utf-8").read(),
        "cache_control": {"type": "ephemeral"}
      }
      
    ],
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "image",
                    "source": {
                        "type": "base64",
                        "media_type": "image/jpeg",
                        "data": encoded_string
                    }
                },
                {
                    "type": "text",
                    "text": "Identify which options have been filled."
                }
            ]
        }
    ]
)

print(message.usage.model_dump_json())

print(message.content[0].text)
```
